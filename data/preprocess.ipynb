```python
# import packages
import pandas as pd # For reading and manipulating 2D data (like spreadsheets)
import numpy as np # For doing numerical calculations (literally NUMerical PYthon)
import matplotlib.pyplot as plt # For making graphs

from sklearn import datasets, linear_model

# Setting to show graphs below cells in colab
%matplotlib inline
```

```python
# CO2
co2 = pd.read_csv(r"../org_data/annual-co2-emissions-per-country.csv")
co2 = co2.loc[co2["Entity"] == "World"]
co2 = co2.filter(items=["Year", "Annual CO₂ emissions"])

co2 = co2.rename(columns={"Annual CO₂ emissions": "CO2"})
co2 = co2.dropna()


co2.to_csv(r"CO2.csv", index=False)

plt.plot(np.asarray(co2["Year"], float), co2["CO2"])
plt.show()
co2

```

```python
# GDP
gdp = pd.read_csv(r"../org_data/global-average-gdp-per-capita-over-the-long-run.csv")
gdp = gdp.loc[gdp["Entity"] == "World"]
gdp = gdp.filter(items=["Year", "GDP per capita"])

gdp = gdp.iloc[15:]     # previous data are not for contiguous years

gdp = gdp.rename(columns={"GDP per capita": "GDP"})
gdp = gdp.dropna()


gdp.to_csv(r"GDP.csv", index=False)

plt.plot(np.asarray(gdp["Year"], float), gdp["GDP"])
plt.show()
gdp
```

```python
ph = pd.read_csv(r"../org_data/climate-change.csv")
ph = ph.loc[ph["Entity"] == "World"]
ph = ph.filter(items=["Date", "Monthly pH measurement"])

ph["Date"] = ph["Date"].map(lambda s: s[:4])
ph = ph.groupby("Date").mean().reset_index()

ph = ph.rename(columns={"Date": "Year", "Monthly pH measurement": "PH"})
ph = ph.dropna()


ph.to_csv(r"PH.csv", index=False)

plt.plot(np.asarray(ph["Year"], float), ph["PH"])
plt.show()
ph
```

```python
# Sea level
sea = pd.read_csv(r"../org_data/sea-level.csv")
sea = sea.loc[sea["Entity"] == "World"]
sea = sea.filter(items=["Day", "Global sea level as an average of Church and White (2011) and UHSLC data"])

sea["Day"] = sea["Day"].map(lambda s: s[:4])
sea = sea.groupby("Day").mean().reset_index()

sea = sea.rename(columns={"Day": "Year", "Global sea level as an average of Church and White (2011) and UHSLC data": "Sea"})
sea = sea.dropna()


sea.to_csv(r"Sea.csv", index=False)

plt.plot(np.asarray(sea["Year"], float), sea["Sea"])
plt.show()
sea
```

```python
# surface temperature
temp = pd.read_csv(r"../org_data/temperature-anomaly.csv")
temp = temp.loc[temp["Entity"] == "Global"]
temp = temp.filter(items=["Year", "Global average temperature anomaly relative to 1961-1990"])

temp = temp.rename(columns={"Global average temperature anomaly relative to 1961-1990": "Temp"})
temp = temp.dropna()


temp.to_csv(r"Temp.csv", index=False)

plt.plot(np.asarray(temp["Year"], float), temp["Temp"])
plt.show()
temp
```
